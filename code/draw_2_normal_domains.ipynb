```python
from IPython.display import HTML
import plotly.graph_objects as go
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
a, b, c, d, e, f = 1, -0.5, 1, 2, 0.1, 1
d1 = np.random.multivariate_normal([0, -4, 0], [[a, 0, b], [0, 0, 0], [b, 0, c]], size=(1000,))
d2 = np.random.multivariate_normal([0, 0, -4], [[d, e ,0], [e, f, 0], [0, 0, 0]], size=(1000,))
```

```python
x = np.linspace(-4, 4, 200)
pdf1x = stats.norm(0, a).pdf(x) * 5 - 4
pdf1z = stats.norm(0, c).pdf(x) * 5 - 4
pdf2x = stats.norm(0, d).pdf(x) * 5 - 4
pdf2y = stats.norm(0, f).pdf(x) * 5 - 4
```

```python
fig = go.Figure(data=[go.Scatter3d(
    x=d1[:, 0],
    y=d1[:, 1],
    z=d1[:, 2],
    mode='markers',
    marker=dict(
        size=2,
        color='blue',                # set color to an array/list of desired values
        # colorscale='Viridis',   # choose a colorscale
        # opacity=0.8
    )
),go.Scatter3d(
    x=d2[:, 0],
    y=d2[:, 1],
    z=d2[:, 2],
    mode='markers',
    marker=dict(
        size=2,
        color='red',                # set color to an array/list of desired values
        # colorscale='Viridis',   # choose a colorscale
        # opacity=0.8
    )
), go.Scatter3d(
    x=x,
    y=pdf1x,
    z=np.ones_like(x) * (-4),
    mode='lines',
    line=dict(
        color='blue',
        width=2
    )
), go.Scatter3d(
    x=np.ones_like(x) * (-4),
    y=pdf1z,
    z=x,
    mode='lines',
    line=dict(
        color='blue',
        width=2
    )
), go.Scatter3d(
    x=np.ones_like(x) * (-4),
    y=x,
    z=pdf2y,
    mode='lines',
    line=dict(
        color='red',
        width=2
    )
), go.Scatter3d(
    x=x,
    y=np.ones_like(x) * (-4),
    z=pdf2x,
    mode='lines',
    line=dict(
        color='red',
        width=2
    )
)
])

# tight layout
fig.update_layout(autosize=False,
                  width=500,
                  height=500,
                  margin=dict(l=0, r=0, b=0, t=0), showlegend=False,
                  scene = dict(
                            xaxis = dict(range=[-4.2, 4.2], visible=True, showticklabels = False, title='',),
                            yaxis = dict(range=[-4.2, 4.2], visible=True, showticklabels = False, title='',),
                            zaxis = dict(range=[-4.2, 4.2], visible=True, showticklabels = False, title='',),
                            camera = dict(center=dict(x=0, y=0, z=-0.15), ),
                            ),
                  )
# fig.show()
HTML(fig.to_html())
```
